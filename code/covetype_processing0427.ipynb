```python
# 2.数据预处理
import pandas as pd

# 假设 df 是您的 DataFrame，它包含上述特征

# 定义一个函数，用于将字节字符串转换为整数
def convert_bytes_to_int(series):
    return series.apply(lambda x: int(x.decode('utf-8')))

# 应用函数到每个二进制特征列
binary_features = ['Wilderness_Area1', 'Wilderness_Area2', 'Wilderness_Area3', 'Wilderness_Area4'] + \
                   ['Soil_Type' + str(i) for i in range(1, 41)]

for feature in binary_features:
    df[feature] = convert_bytes_to_int(df[feature])

# 创建类别标签的映射字典
# 假设 df 是您的 DataFrame，并且它包含 'class' 列

# 创建类别标签的映射字典，使用字节字符串作为键
label_mapping = {b'1': 1, b'2': 2, b'3': 3, b'4': 4, b'5': 5, b'6': 6, b'7': 7}

# 应用映射到 'class' 特征列
df['class'] = df['class'].map(label_mapping)

# 现在 df 中的二进制特征和 'class' 特征都已经被转换为整数了
print(df.info)
```

```python
# 3.合成测试集，保证每个类都有20%数据
import pandas as pd
from sklearn.model_selection import train_test_split

# 假设 df 是你的 DataFrame，并且它有一个名为 'class' 的列

# 首先，我们创建一个字典，它将每个类别的标签映射到其索引
class_to_indices = {cls: df[df['class'] == cls].index for cls in df['class'].unique()}

# 初始化测试集和训练集的索引列表
test_indices = []
train_indices = []

# 遍历每个类别
for cls in df['class'].unique():
    # 选择该类别的所有数据
    class_df = df[df['class'] == cls]

    # 计算该类中应该有多少样本作为测试集
    num_test_samples = int(len(class_df) * 0.2)

    # 从该类数据中随机选择测试集样本
    test_indices_cls = class_df.sample(n=num_test_samples, random_state=42).index

    # 将测试集索引添加到测试集索引列表中
    test_indices.extend(test_indices_cls)

    # 剩余样本的索引作为训练集索引
    train_indices_cls = class_df.drop(test_indices_cls).index

    # 将训练集索引添加到训练集索引列表中
    train_indices.extend(train_indices_cls)

# 使用测试集和训练集的索引来分割 DataFrame
test_df = df.loc[test_indices]
train_df = df.loc[train_indices]


# 保存测试集为 CSV 文件
test_df.to_csv('E:/FedStream/real_data_set/realdataset0427/covertypeNorm/test.csv', index=False)

# 保存训练集为 CSV 文件
train_df.to_csv('E:/FedStream/real_data_set/realdataset0427/covertypeNorm/train.csv', index=False)
```
